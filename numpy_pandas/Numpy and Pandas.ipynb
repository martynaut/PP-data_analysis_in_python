````python
#Groupby
data = {'Firma':['McDonalds', 'McDonalds','KFC','KFC','Starbucks','Starbucks'],
       'Osoba':['Ania','Kasia','Piotr','Mateusz','Basia','Szymon'],
       'Sprzedaż':[400,250,222,333,200,120]}
df = pd.DataFrame(data)
df
````

````python
by_firma.mean()
````

````python
df.groupby('Firma').mean()
````

````python
df_left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df_right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
````

````python
df_left
````

````python
df_right
````

````python
#merge
pd.merge(df_left,df_right,how='inner',on='key')
````

````python
#join
df_left.join(df_right)
````

````python
df_left.join(df_right, rsuffix='right')
````
