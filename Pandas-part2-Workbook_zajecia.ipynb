```python
# wczytaj dane i wypisz 10 pierwszych wierszy
df = pd.read_csv('./data/daily-inmates-in-custody.csv')
df.head(10)
```

```python
# sprawdź typy danych
df.info()
```

```python
# sprawdź wartości występujące w konkretnych kolumnach i zmień typ danych, jeżeli wydaje się to rozsądne 
# usuń puste kolumny
df.drop('DISCHARGED_DT', axis=1, inplace=True)
df.head()
```

```python
df.dropna(inplace=True)
#df['AGE'].fillna(0)
```

```python
df.head()
```

```python
df['AGE'] = df['AGE'].astype(int)
```

```python
df['ADMITTED_DT'] = pd.to_datetime(df['ADMITTED_DT'])
```

```python
df['TOP_CHARGE'] = df['TOP_CHARGE'].str.extract('([0-9]+[\.]{0,1}[0-9]*)').astype(float)
```

```python
df['TOP_CHARGE'].fillna(0, inplace=True)
```

```python
df['year'] = df['ADMITTED_DT'].dt.year
```

```python
df.head()
```

```python
df.groupby('year')[['TOP_CHARGE']].mean()
```

### apply i lambda functions

```python
def change_gender(x):
    if x== 'female':
        return 'F'
    elif x=='male':
        return 'M'

# zmodyfikuj wartość w w kolumnie 'GENDER'
# 'F' - 'female' i 'M' - 'male'
# df['GENDER'].str.replace('F', 'female').str.replace('M', 'male')
# df.loc[df['GENDER']=='F', 'GENDER'] = 'female'
# df.loc[df['GENDER']=='M', 'GENDER'] = 'male'
df['GENDER'].apply(lambda x: len(x)+len(x)*2)
```

```python
df['GENDER'].unique()
```

```python
def funkcja_1(x):
    return x['INMATE_STATUS_CODE']+'_'+x['GENDER']

df.apply(funkcja_1, axis=1)
```

```python
# wykonaj z wykorzystaniem funkcji apply
```

```python
# połącz dwie kolumn CUSTODY_LEVEL i RACE jako MED_RACE, jeżeli którakolwiek wartość jest nieznana wpisz 'unknown'

def funkcja_1(x):
    if not pd.isnull(x['CUSTODY_LEVEL']) and not pd.isnull(x['RACE']):
        return x['CUSTODY_LEVEL']+'_'+x['RACE']
    else:
        return 'unknown'

df['MED_RACE'] = df.apply(funkcja_1, axis=1)
```
