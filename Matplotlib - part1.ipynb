## Zadanie 1. zróbcie analogiczny wykres dla n_injured

# Łączenie wykresów

```python
df_by_year_inj = df.groupby('year')[['n_injured']].sum().reset_index()
```

```python
plt.plot(df_by_year['year'], df_by_year['n_killed'], label='killed')
plt.plot(df_by_year_inj['year'], df_by_year_inj['n_injured'], label='injured')

plt.ylim(0,40000)
plt.xlim(2013, 2018)
plt.title('People killed/injured per year in gun shoot in US')
plt.xlabel('year')
plt.ylabel('# people')
plt.legend()
# plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(df_by_year['year'], df_by_year['n_killed'], label='killed', color='red')
ax2 = ax.twinx()
ax2.plot(df_by_year_inj['year'], df_by_year_inj['n_injured'], label='injured', color='blue')

plt.xlim(2013, 2018)
plt.title('People killed/injured per year in gun shoot in US')
ax.set_xlabel('year')
ax.set_ylabel('# people killed')
ax2.set_ylabel('# people injured')

ax.legend(loc=0)
ax2.legend(loc=1)

# plt.show()
```

## Dwa wykresy obok siebie

```python
fig, axes = plt.subplots(1,2, figsize=(15,10))


axes[0].plot(df_by_year['year'], df_by_year['n_killed'], label='killed')
axes[0].set_xlim(2013, 2018)
axes[0].set_title('People killed/injured per year in gun shoot in US')
axes[0].set_xlabel('year')
axes[0].set_ylabel('# people')
axes[0].legend()


axes[1].plot(df_by_year_inj['year'], df_by_year_inj['n_injured'], label='injured')
axes[1].set_xlim(2013, 2018)
axes[1].set_title('People killed/injured per year in gun shoot in US')
axes[1].set_xlabel('year')
axes[1].set_ylabel('# people')
axes[1].legend()
# plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize=(15,10), sharex=True, sharey=True)


axes[0].plot(df_by_year['year'], df_by_year['n_killed'], label='killed')
axes[0].set_xlim(2013, 2018)
axes[0].set_title('People killed per year in gun shoot in US')
axes[0].set_xlabel('year')
axes[0].set_ylabel('# people')
axes[0].legend()


axes[1].plot(df_by_year_inj['year'], df_by_year_inj['n_injured'], label='injured')
axes[1].set_title('People injured per year in gun shoot in US')
axes[1].set_ylabel('# people')
axes[1].legend()
# plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(df_by_year['year'], df_by_year['n_killed'], label='killed', color='red')
ax.set_xlim(2013, 2018)
ax.set_title('People killed/injured per year in gun shoot in US')
ax.set_xlabel('year')
ax.set_ylabel('# people killed')
ax.legend(loc=0)
ax.set_yscale('log')

ax.set_xticks(df_by_year['year'], ['rok '+ str(x) for x in df_by_year['year']])

ax.text(2016.5, 10000, 'Maximum')
# plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(df_by_year['year'], df_by_year['n_killed'], label='killed', color='red')
ax.set_xlim(2013, 2018)
ax.set_title('People killed/injured per year in gun shoot in US')
ax.set_xlabel('year')
ax.set_ylabel('# people killed')
ax.legend(loc=0)
ax.set_xticks(df_by_year['year'], ['rok '+ str(x) for x in df_by_year['year']])


yr_max = df_by_year.iloc[df_by_year['n_killed'].argmax()]['year']
nk_max = df_by_year['n_killed'].max()
# Add a black arrow annotation
plt.annotate('Maximum', xy=(yr_max, nk_max), xytext=(yr_max-0.5, nk_max-5000), arrowprops=dict(facecolor='black'))
# plt.show()
```
